```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
arr_x = np.linspace(1, 10, 20)
```

```python
arr_x
```

```python
arr_y = arr_x**2
arr_y
```

```python
# Functional methods
plt.plot(arr_x,arr_y)
plt.title("Example")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(arr_x,arr_y, color='r',lw=5, ls="-.")

plt.subplot(1,2,2)
plt.plot(arr_y,arr_x)
```

```python
# OOP
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(arr_x, arr_y)
axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")
axes.set_title("Example")
```

```python
#Subplot

fig2 = plt.figure()

# Big graph
axes_1 = fig2.add_axes([0,0,1,1])
axes_1.plot(arr_x,arr_y)

# Small graph
axes_2 = fig2.add_axes([0.2, 0.3, 0.5, 0.5]) #x-offset, y-offset, width, height
axes_2.plot(arr_y, arr_x, alpha=0.5)
```
